# Task 4 Creating a loyalty flag for existing customers

### Creating max_order column

```python
#creating max order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['clients_order_number'].transform(np.max)
```

```python
#check max_order clmn
ords_prods_merge.head(100)
```

### creating a loyalty flag column

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#value count loyalty customer
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head(10)
```

# Task 5 Check the basic statistics of the product prices for each loyalty category 

```python
#calculating product prices for loyal customers
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

# Task 6 Create a spending flag

### Creating Mean column

```python
### creating max order clm for prices
ords_prods_merge['mean_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check mean_order clmn
ords_prods_merge.head(10)
```

### Creating Spending flag clmn

```python
ords_prods_merge.loc[ords_prods_merge['mean_order'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#value count Spending flag
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

# Task 7 Creating order freq. flag

### Creating Freq column for days since prior order

```python
# creating max order clm for prices
ords_prods_merge['frq_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check frq_order clmn
ords_prods_merge.head(30)
```

### Creating Freq flag

```python
ords_prods_merge.loc[ords_prods_merge['frq_days_since_prior_order'] <= 10, 'FRQ_prior_order_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['frq_days_since_prior_order'] <= 20) & (ords_prods_merge['frq_days_since_prior_order'] > 10), 'FRQ_prior_order_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['frq_days_since_prior_order'] > 20, 'FRQ_prior_order_flag'] = 'Non-frequent customer'
```

```python
#value count Spending flag
ords_prods_merge['FRQ_prior_order_flag'].value_counts(dropna=False)
```

```python
#creating sub-set to evaluate nan values
df_prior_order_frq= ords_prods_merge[ords_prods_merge['FRQ_prior_order_flag'].isnull() == True]
```
